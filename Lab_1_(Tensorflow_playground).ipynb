## Part 1

```
from PIL import Image, ImageDraw
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline

import networkx as nx

n = np.random.randint(5,50)
p = np.random.rand()**2 # square to bias lower edge densities
node_sizes = np.random.randint(50,500,size=n)
colors = np.random.rand(n)

g = nx.gnp_random_graph(n,p)
nx.draw(g,node_size=node_sizes, pos=nx.spring_layout(g),
        cmap='viridis', vmin=0.0, vmax=1.0, node_color=colors)
title = "Erdos-Renyi random graph with " + str(n)
title += " vertices and edge probability " +str(np.round(p,2))
plt.title(title)
plt.show()
```
